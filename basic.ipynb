```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Example Data
X = [1, 3, 4, 7, 9, 12]
y = [12, 11, 14, 23, 9, 13]
```

```python
# Create figure containing a single axis
fig, ax = plt.subplots()
ax.plot(X, y)
```

```python
# Figure means a graphical board where the plots can 
fig_1, (ax_1, ax_2) = plt.subplots(2, 1)

ax_1.set_title("Sample Xy plot")
ax_1.set_title("Sample Xy plot (with y's elements increased by 5) ")
ax_1.plot(X, y)
ax_2.plot(X, [(i + 5) for i in y])
```

```python
# Create figures to compare the linear, quadratic and cubic equations (distinguished between figs)
x = np.linspace(0, 3, 100)
fig, (ax_lin, ax_quad, ax_cub) = plt.subplots(3, 1,
                                             figsize=(10, 9),
                                             layout='constrained')

ax_lin.plot(x, x, label='linear')
ax_quad.plot(x, x**2, label='quadratic')
ax_cub.plot(x, x**3, label='cubic')
```

```python
# Create figures to compare the linear, quadratic and cubic equations (distinguished between figs)
x = np.linspace(0, 15, 100)
fig, ax = plt.subplots(1, 1,
                      figsize=(12, 5),
                      layout='constrained')

ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Comparison among 3 type of equations')
ax.legend()
```

```python
X = pd.read_csv("./data/v4_Latest_Data_Science_Salaries.csv")
salary = X["Salary"].to_numpy()
id = [i for i in range(0, len(salary))]

fig, ax = plt.subplots(1, 1,
                       layout='constrained')

ax.plot(id[:100], salary[:100], label='Salary 0-99')
ax.plot(id[:100], salary[100:200], label="Salary 100-199")
ax.plot(id[:100], [salary[:100].sum()/len(salary[:100]) for _ in range(100)], label="Mean 0-99")
ax.plot(id[:100], [salary[100:200].sum()/len(salary[100:200]) for _ in range(100)], label="Mean 100-199")
ax.set_title("Salary among workers working in the Data Industry")
ax.set_xlabel("Worker")
ax.set_ylabel("Salary (USD)")

ax.legend()
```
